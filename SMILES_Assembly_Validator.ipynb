```python
# ========== CELL 1 (manual fragment assembly, zero copy-paste) ==========
!pip install -q rdkit

from rdkit import Chem
from rdkit.Chem import Descriptors

# Type each fragment manually (short, simple, no special chars that corrupt)
fragments = [
    "C[C@H]1",
    "C(=O)N(",      # O is capital letter O, not zero
    "c2cc(F)ccn2",
    ")C2=C1",
    "C(=O)N(C2)",
    "C(C)C(=O)",
    "N3CC",
    "[C@H](C3)",
    "C(=O)Nc4ccc(",
    "CCO",          # PEG spacer, piece 1
    "CCO",          # PEG spacer, piece 2
    "CCOc5ccc(",
    "C=Cc6cc(Cl)ccc7nc(",
    "C(=O)O",       # Carboxylic acid
    ")cc67)",
    "cc5)",
    "cc4"
]

# Assemble in code (you never touch the full string)
smiles = "".join(fragments)
print("Assembled SMILES length:", len(smiles))
print("SMILES:", smiles)

# Parse
mol = Chem.MolFromSmiles(smiles)
if mol is None:
    print("❌ Parse failed – check fragments 3, 9, 10, 11 for typos")
else:
    Chem.SanitizeMol(mol)
    print("✅ Success!")
    print("Heavy atoms:", mol.GetNumAtoms())
    print("MW:", Descriptors.MolWt(mol))
    print("clogP:", Descriptors.MolLogP(mol))
```

```python
smiles = "".join(fragments)
print("Full SMILES:", smiles)
print("Length:", len(smiles))

# Count parentheses balance
open_parens = smiles.count("(")
close_parens = smiles.count(")")
print(f"Parentheses: {open_parens} open, {close_parens} close")

# Count ring closures
ring_nums = set(c for c in smiles if c.isdigit())
print(f"Ring closures used: {ring_nums}")

mol = Chem.MolFromSmiles(smiles)
if mol is None:
    print("❌ Parse failed")
else:
    print(f"✅ Parsed as {mol.GetNumAtoms()} atoms")
    # Check if it's actually one molecule or multiple fragments
    from rdkit.Chem import rdmolops
    num_fragments = len(rdmolops.GetMolFrags(mol))
    print(f"Number of disconnected fragments: {num_fragments}")
```
